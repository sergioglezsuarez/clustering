```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datos_practica10.csv")
df
```

```python
for i in range(1, 9):
    print("KMeans {} clusters".format(i))
    kmeans = KMeans(n_clusters=i, random_state=42).fit(df)
    print(kmeans.labels_)
    fig, ax = plt.subplots()
    ax.scatter(df["x1"], df["x2"], c=kmeans.labels_)
    plt.show()
    print("AgglomerativeClustering 'average' {} clusters".format(i))
    agg1 = AgglomerativeClustering(n_clusters=i, linkage='average').fit(df)
    print(agg1.labels_)
    fig, ax = plt.subplots()
    ax.scatter(df["x1"], df["x2"], c=agg1.labels_)
    plt.show()
    print("AgglomerativeClustering 'single' {} clusters".format(i))
    agg2 = AgglomerativeClustering(n_clusters=i, linkage='single').fit(df)
    print(agg2.labels_)
    fig, ax = plt.subplots()
    ax.scatter(df["x1"], df["x2"], c=agg2.labels_)
    plt.show()
```

En primer lugar, como es obvio, al usar 1 cluster, en los tres casos se obtiene el mismo resultado, que todas las muestras pertenecen a este Ãºnico cluster.

Sin embargo, a medida que se aumenta el nÃºmero de clusters, observamos como KMeans y AgglomerativeClustering con linkage 'average' producen resultados muy parecidos hasta llegar a 5 clusters. De 5 a 7 clusters el mejor resultado se obtiene con KMeans, pero, curiosamente, con 8 clusters, comparando estos dos mÃ©todos, parece que AgglomerativeClustering hace un mejor trabajo, ya que visualmente los clusters estÃ¡n mÃ¡s claramente separados.

En cuanto a AgglomerativeClustering con linkage 'single', los nuevos clusters que se van formando son de una o muy pocas muestras (_n clusters=6_), por lo que si quisiÃ©ramos realizar un clustering de pocos clusters, este mÃ©todo no es eficaz para este conjunto de datos, ya que las muestras estÃ¡n relativamente juntas.

```python
silhouettes=[]
for i in range(2, 11):
    print("KMeans {} clusters".format(i))
    kmeans = KMeans(n_clusters=i, random_state=42).fit(df)
    silhouettes.append(silhouette_score(df, kmeans.labels_, random_state=42))
    print(silhouette_score(df, kmeans.labels_, random_state=42))
    fig, ax = plt.subplots()
    ax.scatter(df["x1"], df["x2"], c=kmeans.labels_)
    plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(list(range(2, 11)), silhouettes)
plt.xlabel("NÃºmero de clusters")
plt.ylabel("Valor medio de la silueta")
plt.show()
```
